```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

dfx = pd.read_csv('Training Data/Linear_X_Train.csv')
dfy = pd.read_csv('Training Data/Linear_Y_Train.csv')
x = dfx.values
y = dfy.values
x = x.reshape((-1,1))
y = y.reshape((-1,1))
z=LinearRegression()

```

```python
X = (x-x.mean())/x.std()
Y = y
```

```python
plt.scatter(X,Y)
plt.show()
```

```python
z.fit(X,Y)
```

```python
predicted=z.predict(dfx)
print(predicted)
```

```python
plt.plot(dfx,predicted,color="yellow")
plt.scatter(dfx,dfy,color="blue")
plt.show()
```

```python
x_test=pd.read_csv("Testing Data/Linear_X_Test.csv")
X1=x_test.values
X1=X1.reshape((-1,1))
X1=(X1-X1.mean())/X1.std()
z1=z.predict(X1)
```

```python
plt.scatter(X,Y,color="blue")
plt.plot(x_test,z1,color="yellow",label="prediction")
plt.legend()
plt.show()
```

```python
bias = z.intercept_
coeff = z.coef_
print(bias)
print(coeff)
```

```python
z.score(X,Y)
```
